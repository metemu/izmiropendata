```python
import requests
import folium
import urllib
from pandas.io.json import json_normalize
import pandas as pd
import folium
import geopandas
from geopandas import GeoDataFrame
from shapely.geometry import Point
from pyproj import Proj
```

```python
#verify=False ile SSL error atlamış oluyoruz
saglik1 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/liseler'
saglik2 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/meslekliseleri'
saglik3 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/kolejler'
saglik4 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/ortaokullar'
saglik5 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/ilkokullar'
saglik6 = 'https://openapi.izmir.bel.tr/api/ibb/cbs/anaokullari'

data1 = requests.get(saglik1, verify=False)
data2 = requests.get(saglik2, verify=False)
data3 = requests.get(saglik3, verify=False)
data4 = requests.get(saglik4, verify=False)
data5 = requests.get(saglik5, verify=False)
data6 = requests.get(saglik5, verify=False)
```

```python
df1 = pd.DataFrame(data1.json())
df1 = pd.DataFrame(df1['onemliyer'].tolist(), index=df1.index)

df2 = pd.DataFrame(data2.json())
df2 = pd.DataFrame(df2['onemliyer'].tolist(), index=df2.index)

df3 = pd.DataFrame(data3.json())
df3 = pd.DataFrame(df3['onemliyer'].tolist(), index=df3.index)

df4 = pd.DataFrame(data4.json())
df4 = pd.DataFrame(df4['onemliyer'].tolist(), index=df4.index)

df5 = pd.DataFrame(data5.json())
df5 = pd.DataFrame(df5['onemliyer'].tolist(), index=df5.index)

df6 = pd.DataFrame(data6.json())
df6 = pd.DataFrame(df6['onemliyer'].tolist(), index=df6.index)

df = pd.concat([df1, df2, df3, df4, df5, df6])
df
```

```python
geometry = [Point(xy) for xy in zip(df['BOYLAM'], df['ENLEM'])]
gdf = GeoDataFrame(df, crs=4326, geometry=geometry)
gdf.head()
```

```python
m = folium.Map(location=[38.5, 27], zoom_start=8)
folium.GeoJson(gdf).add_to(m)
m
```

```python
# Get the latitude and longitude from result as a list
locations = gdf[['ENLEM', 'BOYLAM']]
coords = locations.values.tolist()

from folium.plugins import MarkerCluster

m = folium.Map([38.5, 27], zoom_start= 8, tiles='cartodbpositron')

marker_cluster = MarkerCluster().add_to(m)

for point in range(0, len(coords)):
    folium.Marker(location = coords[point], popup= 'Ad: ' + str(df['ADI'][point])).add_to(marker_cluster)

m
```
